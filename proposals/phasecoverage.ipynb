# Phase coverage in planet searches

### Assumptions 
 * We have time series observations of a target that may host a planet
 * We're interested in getting observations that completely cover orbital periods of 8-10 hours
 * Our exposures are taken in one minute cadence
 * Transits are very deep and S/N is a non-issue - if you observe during transit, you will detect it (yeah, this is a big assumption)

### Quick dumb algorithm

Given a time series $t$, the orbital phases $\phi$ sampled for a given orbital period $P_{test}$ are

$$ \phi = \text{mod}\left( t/P_{test}, 1 \right). $$

The widest range of phases not covered by the time series would be

$$ \phi = \text{sort}(\phi) $$
$$ \Delta\phi_{max} = \text{max} \left( \phi_{i+1} - \phi_{i} \right )$$

and we need to add the special case for the difference between the maximum and minimum phase

$$ \Delta\phi_{max} = \text{max} \left( \phi_{i+1} - \phi_{i},~\text{min}(\phi) - (\text{max}(\phi) - 1)\right ).$$

When the duration of transit $\tau_{14} > P_{test} \times \Delta \phi_{max}$, we have "complete coverage" of the test period $P_{test}$. Iterate over all $P_{test}$ of interest. 

```python
%pylab inline

import numpy as np
from matplotlib import pyplot as plt

# Sample times: 1 day worth of seconds
Nminutesinday = 24*60.0
minutes = np.arange(2*Nminutesinday)

# Make fake fluxes
fluxes = np.ones_like(minutes) + 0.05*np.random.randn(len(minutes))

# Make a gap in those times
takeoutgap = (minutes < 1.0*Nminutesinday)*(minutes > 0.5*Nminutesinday) + \
             (minutes < 2.0*Nminutesinday)*(minutes > 1.5*Nminutesinday)
minutes_gap = minutes[takeoutgap]
fluxes_gap = fluxes[takeoutgap]
fig, ax = plt.subplots(1, figsize=(14,5))
ax.plot(minutes_gap/60.0, fluxes_gap,'.')
ax.set_title('Fake incomplete data')
ax.set_xlabel('Time (hours')
ax.set_ylabel('Flux')
ax.set_xlim([0,48])
plt.show()
```

```python
testperiods = np.linspace(60, np.max(minutes_gap) - np.min(minutes_gap), 100)#[Nminutesinday-600] # In minutes    np.arange(0.8,)

minimumduration = 1.0 # minutes
incompletetolerance = 0.5 # minutes

completeness = np.zeros(len(testperiods))
for i, testperiod in enumerate(testperiods):
    phasescovered = minutes_gap/testperiod % 1
    largestphasegap = np.max(np.diff(np.sort(phasescovered)))
    
    ## Also check for gap between first and last points
    largestphasegap = np.max([largestphasegap,\
                         np.min(phasescovered) - (np.max(phasescovered) - 1)])
    largesttimegap = largestphasegap*testperiod # minutes

    if np.abs(largesttimegap - minimumduration) < incompletetolerance:
        #print 'Complete for periods of %.2f min = %.2f hours' % (testperiod, testperiod/60.0)
        completeness[i] = 1.0
    #else: 
    #    print 'Largest phase gap = %.3f = %.3f min = %.3f days' % (largestphasegap, largesttimegap, largesttimegap/(24*60))

fig, ax = plt.subplots(1, figsize=(14,6))
barwidth = (testperiods[1]-testperiods[0])/60.0
ax.bar(testperiods/60., completeness, barwidth, lw=2, color='k')
ax.axvline(8, ls=':', color='red')
ax.axvline(10, ls=':', color='red')
ax.set_xlabel('Test Period (hours)')
ax.set_ylabel('Completeness')
ax.set_title('Phase Coverage')
plt.show()
```
