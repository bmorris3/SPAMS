# Querying WD Catalogs

[Holberg et al. 2008](http://adsabs.harvard.edu/abs/2008AJ....135.1225H) collected 122 WDs within 20 pc. 

```python
%pylab inline
import numpy as np
from matplotlib import pyplot as plt
import ephem

def floatorzero(string):
    if len(string) > 0:
        return float(string)
    else:
        return 0.0

rawtxt = open('holberg2008.txt','r').read().splitlines()
Nheaderrows = 47
Nelements = len(rawtxt) - Nheaderrows
RA_deg = np.zeros(Nelements)  # RAJ2000 in degrees
Dec_deg = np.zeros(Nelements) # DecJ2000 in degrees
temp = np.zeros(Nelements)
V = np.zeros(Nelements)
binary = Nelements*['']#np.zeros(Nelements)
for i in range(Nelements):
    splitline = rawtxt[Nheaderrows + i].split('\t')
    RA_deg[i] = float(splitline[0])
    Dec_deg[i] = float(splitline[1])
    V[i] = float(splitline[5])
    binary[i] = splitline[6].strip()

RA_str = [str(ephem.hours(ephem.degrees(RA_deg_i*np.pi/180))) for RA_deg_i in RA_deg]
Dec_str = [str(ephem.degrees(Dec_deg_i*np.pi/180)) for Dec_deg_i in Dec_deg]
```

# How bright and where are the WDs?

```python
nonbinary = np.array([True if len(binary[i]) == 0 else False for i in range(len(binary))])
print '%i stars not in binaries.' % (np.sum(nonbinary))
fig, ax = plt.subplots(1,2,figsize=(18,8))
ax[0].plot(RA_deg[nonbinary], Dec_deg[nonbinary], '.')
ax[0].set_xlabel('RA')
ax[0].set_ylabel('Dec')
ax[0].set_title('Sky Map')

Nbins = 20
ax[1].hist(V, bins=Nbins, histtype='stepfilled', fc='silver')
ax[1].set_title('V Magnitudes')
ax[1].set_xlabel('Magnitude')
ax[1].set_ylabel('Number of WDs')
ax[1].legend()
plt.show()
```

# How many are there and where are brightest stars?

```python
bright_nb = (V < 13)*nonbinary

fig, ax = plt.subplots(1,2,figsize=(18,8))
ax[0].plot(RA_deg[bright_nb], Dec_deg[bright_nb], '.')
ax[0].set_xlabel('RA')
ax[0].set_ylabel('Dec')
ax[0].set_title('Sky Map')

Nbins = 20
ax[1].hist(V[bright_nb], bins=Nbins, histtype='stepfilled', fc='silver')
ax[1].set_title('V Magnitudes')
ax[1].set_xlabel('Magnitude')
ax[1].set_ylabel('Number of WDs')
ax[1].legend()
plt.show()
```

# McCook et al. 1999 with V < 14 and only non-binaries
[McCook et al. 1999](http://adsabs.harvard.edu/cgi-bin/bib_query?1999ApJS..121....1M) is one of the biggest lists.

```python
%pylab inline
import numpy as np
from matplotlib import pyplot as plt
import ephem

def floatorzero(string):
    if len(string) > 0:
        return float(string)
    else:
        return 0.0

rawtxt = open('mccook1999.txt','r').read().splitlines()
Nheaderrows = 63
Nelements = len(rawtxt) - Nheaderrows
RA_deg = np.zeros(Nelements)  # RAJ2000 in degrees
Dec_deg = np.zeros(Nelements) # DecJ2000 in degrees
temp = np.zeros(Nelements)
V = np.zeros(Nelements)
binary = Nelements*['']#np.zeros(Nelements)
names = Nelements*['']
for i in range(Nelements):
    splitline = rawtxt[Nheaderrows + i].split('\t')
    if len(splitline[0].strip()) > 1:
        RA_deg[i] = float(splitline[0])
        Dec_deg[i] = float(splitline[1])
        names[i] = splitline[7].strip()
        V[i] = float(splitline[8])
        binary[i] = splitline[6].strip()

#nonbinary = np.array([True if sptype.count('+') == 0 else False for i in range(len(binary))])
nonbinary = np.array([True if binary[i] == '' else False for i in range(len(binary))])
RA_str = [str(ephem.hours(ephem.degrees(RA_deg_i*np.pi/180))) for RA_deg_i in RA_deg]
Dec_str = [str(ephem.degrees(Dec_deg_i*np.pi/180)) for Dec_deg_i in Dec_deg]
```

```python
nonzero_nb = (V != 0)*nonbinary
print '%i non-binary WDs with V < 14.' % (np.sum(nonzero_nb))

fig, ax = plt.subplots(1,2,figsize=(18,8))
ax[0].plot(RA_deg[nonzero_nb], Dec_deg[nonzero_nb], '.')
ax[0].set_xlabel('RA')
ax[0].set_ylabel('Dec')
ax[0].set_ylim([-40,90])
ax[0].set_title('Sky Map')

Nbins = 15
ax[1].hist(V[nonzero_nb], bins=Nbins, histtype='stepfilled',fc='w')
ax[1].set_title('V Magnitudes')
ax[1].set_xlabel('Magnitude')
ax[1].set_ylabel('Number of WDs')
#ax[1].set_xlim([np.min([u[nonzerovalues].min(), g[nonzerovalues].min(), r[nonzerovalues].min()]), 21])
plt.show()
```
