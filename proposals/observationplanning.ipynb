```python
%pylab inline
import ephem
import numpy as np
from matplotlib import pyplot as plt
import datetime 
import matplotlib
matplotlib.rcParams['font.size'] = 13
# Set up APO observatory
observatory_minHorizon = ephem.degrees('25:00:00')
apo = ephem.Observer()
apo.lat = '32.0:46.0:49.0' 
apo.long = '-105.0:49.0:13.0' 
apo.elevation = 2788.0 # m
apo.temp = 10.0 ## Celsius
apo.horizon = observatory_minHorizon

# Set up stars
KUV = ephem.FixedBody()
KUV._ra = ephem.hours('08:40:08.2')  
KUV._dec = ephem.degrees('+40:15:04')

#WD1150153._ra = ephem.hours('11:53:15.272')
#WD1150153._dec = ephem.degrees('-15:36:36.60')

WD1307354 = ephem.FixedBody() 
WD1307354._ra = ephem.hours('13:09:57.69')
WD1307354._dec = ephem.degrees('+35:09:47.1')

HAT25 = ephem.FixedBody()
HAT25._ra = ephem.hours('03:13:44.500')
HAT25._dec = ephem.degrees('+25:11:50.68')

apo.date = ephem.date(datetime.datetime(2015,2,10,10))
apo.horizon = ephem.degrees('-06:00:00')
sunrise = apo.next_rising(ephem.Sun()).datetime()
sunset = apo.previous_setting(ephem.Sun()).datetime()
apo.horizon = observatory_minHorizon
KUValtitudes = []
WD1307354altitudes = []
HAT25altitudes = []
dates = []
counter = 0
while sunset + datetime.timedelta(hours=counter) < sunrise:
 time = sunset+datetime.timedelta(hours=counter)
 apo.date = ephem.date(time)
 dates.append(time)
 KUV.compute(apo)
 WD1307354.compute(apo)
 HAT25.compute(apo)
 counter += 0.1
 KUValtitudes.append(KUV.alt*180/np.pi)
 WD1307354altitudes.append(WD1307354.alt*180/np.pi)
 HAT25altitudes.append(HAT25.alt*180/np.pi)

#halftime = datetime.datetime(2014,11,9,23,4)+datetime.timedelta(hours=7)
halftime = sunset + datetime.timedelta(seconds=0.5*(sunrise-sunset).total_seconds())
fig, ax = plt.subplots(figsize=(14,10))
ax.plot_date(dates, KUValtitudes, lw=2, fmt='-', label='KUV 08368+4026')
ax.plot_date(dates, WD1307354altitudes, lw=2, fmt='-', label='WD 1307+354')
ax.plot_date(dates, HAT25altitudes, lw=2, fmt='-', label='HAT-P-25')

ax.axhline(7, color='k', ls='--', label='Pointing limit')
ax.axvline(halftime, color='k', lw=2, label='Half night')

ax.set_ylim([0,90])
#ax.set_xlim([sunset,halftime])
ax.set_ylabel('Altitude (degrees)')
ax.set_xlabel('Time beginning with sunset: %s' % str(sunset).split('.')[0])
ax.set_title('APO: %s (UT)' % sunset.strftime('%b %d, %Y'))
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(14,10))
KUV_am = 1./np.sin(np.array(KUValtitudes)*np.pi/180)
WD1307354_am = 1./np.sin(np.array(WD1307354altitudes)*np.pi/180)
HAT25_am = 1./np.sin(np.array(HAT25altitudes)*np.pi/180)

for am in [KUV_am, WD1307354_am, HAT25_am]:
 am[am < 1] = 10
 
ax.plot_date(dates, KUV_am, lw=2, fmt='-', label='KIC 9652632')
ax.plot_date(dates, WD1307354_am, lw=2, fmt='-', label='WD 1307+354')
ax.plot_date(dates, HAT25_am, lw=2, fmt='-', label='HAT-P-25')
ax.axhline(2, color='k', ls='--')

ax.set_ylim([1,3])
#ax.set_xlim([sunset,halftime])
ax.set_ylabel('Airmass')
ax.set_xlabel('Time beginning with sunset: %s' % str(sunset).split('.')[0])
ax.set_title('APO: November 9, 2014, A half')
ax.legend()
plt.show()
```
