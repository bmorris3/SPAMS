# A practical introduction to statistics and uncertainty in physics

Let's say you wanted to measure the rate at which water was falling during a good rain. You place a bucket on a scale and place it outside. You want to measure the rate of rainfall in kilograms of water that fall within your bucket per hour. 

Your scale works just fine, but the measurements of weight that it takes are imperfect as all detectors are. If you measure the weight of a bucket 50 times when it is not raining, you see weight measurements like this:

```python
%pylab inline
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
plt.style.use('ggplot')
matplotlib.rcParams['font.size'] = 14

trial = np.arange(0,50)
bucketweight = 10*np.random.randn(50) + 500
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(trial, bucketweight, 'ko')
ax.set_xlabel('Measurement Trial ($i$)')
ax.set_ylabel('Bucket Weight (grams)')
ax.set_title('Bucket weight - no rain')
plt.show()
```

There's a good deal of scatter in that weight measurements that comes from the process of measurement itself. The weight of the bucket is not fluctuating but our measurements of the weight fluctuates. If we want to state what the weight of the bucket is, we might take an average of all of the weights which we will call $\bar{w}$. If we represent each weight measurement as $w_i$, which we measured $N$ times, the mean weight is

$$ \bar{w} = \frac{1}{N}\sum_{i=0}^N w_i$$

but we know that there's some uncertainty associated with that mean measurement -- how do we quantify it?

# Standard Deviation

One way to state the uncertainty is to take the standard deviation of the measurements $\sigma$, which looks like this

$$ \sigma = \sqrt{ \frac{1}{N}\sum_{i=0}^N \left( \bar{w} - w_i \right)^2 }$$. 

What is this equation doing? Inside the parentheses we're taking the square of the difference between the average weight $\bar{w}$ and a particular weight measurement $w_i$ which tells us how far that particular weight measurement was from the mean. Then we sum this distance-from-the-mean metric up over all weight measurements and take the square root to give us the standard deviation $\sigma$. Using numpy, you can compute the standard deviation with `numpy.std()`.

We can use $\sigma$ as a means of describing how broad the scatter is, since if the points all far fall from the mean the standard deviation is large, if the points fall near to the mean, the standard deviation is small:

```python
bigscatter = np.random.randn(100)
smallscatter =  0.4*np.random.randn(100)

fig, ax = plt.subplots(1, 2, figsize=(16,6), sharey=True)
ax[0].plot(smallscatter, 'bo', label='$w_i$')
ax[0].axhline(np.mean(smallscatter), color='k', ls='--', label=r'$\bar{w}$')
for coeff, label in zip([1, -1], [r'$\bar{w}\pm\sigma$', '']):
    ax[0].axhline(np.mean(smallscatter) + coeff*np.std(smallscatter), color='r', ls='--', label=label)
ax[0].set_title('Small scatter: $\sigma = %f$' % np.std(smallscatter))
ax[0].legend(numpoints=1)

ax[1].plot(bigscatter, 'go', label='$w_i$')
ax[1].set_title('Big scatter: $\sigma = %f$' % np.std(bigscatter))
ax[1].axhline(np.mean(bigscatter), color='k', ls='--', label=r'$\bar{w}$')
for coeff, label in zip([1, -1], [r'$\bar{w}\pm\sigma$', '']):
    ax[1].axhline(np.mean(bigscatter) + coeff*np.std(bigscatter), color='r', ls='--', label=label)
ax[1].legend(numpoints=1)

plt.show()
```

Going back to our bucket - let's say you have *a lot* of weight measurements. Like this:

```python
trial = np.arange(0,1e4)
bucketweight = 10*np.random.randn(1e4) + 500
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(trial, bucketweight, 'k.')
ax.set_xlabel('Measurement Trial ($i$)')
ax.set_ylabel('Bucket Weight (grams)')
ax.set_title('Bucket weight - no rain')
plt.show()
```

Let's imagine sitting where the label is on the $y$ axis and looking across the plot through the forest of points -- we'd see something like this histogram of the number of measurements that fall in a given range of weights:

```python
fig, ax = plt.subplots(figsize=(10,6))
n, bins, patches = ax.hist(bucketweight, 50, color='k')
ax.set_ylabel('Number of measurements')
ax.set_xlabel('Weight (grams)')
for coeff, label in zip([1, -1], [r'$\bar{w}\pm\sigma$', '']):
    ax.axvline(np.mean(bucketweight) + coeff*np.std(bucketweight), color='r', ls='--', lw=3, label=label)
ax.axvline(np.mean(bucketweight), color='w', ls='--', lw=4, label=r'$\bar{w}$')
ax.legend()
plt.show()
```

If you have random errors on each measurement, as our scale does, then we will see this characteristic [normal distribution](http://en.wikipedia.org/wiki/Normal_distribution) in the histogram. 

# Probability and $\sigma$

Let's connect these ideas to probability now. Let's say you were to take another measurement of the weight of the bucket. Looking at the histogram above, you can see that most of the time the weight comes out to be between about 490 and 510 grams, so it's pretty likely that you'll measure a weight in that range again. You'd expect that it's less likely by comparison that you'd measure the weight to be between 510 to 520, and even less likely that you'd measure a weight in the range of 530-540. 

We can connect these likelihoods back to the standard deviation because 68.3% of the measurements that we take will fall within $\sigma$ of the mean. Practically speaking, let's check this in our example -- what percentage of our points fall within $\sigma$ of $\bar{w}$?

```python
print 'Points within one-sigma: %.1f%%' % (np.sum((bucketweight < np.mean(bucketweight) + np.std(bucketweight))*\
                                                  (bucketweight > np.mean(bucketweight) - np.std(bucketweight)))*\
                                           100.0/len(bucketweight))
```

The probability that a measurement will fall within a certain number of sigma of the mean is given by this function ([cumulative distribution](http://en.wikipedia.org/wiki/Standard_deviation#Rules_for_normally_distributed_data)):

```python
from scipy.stats import norm
Nsigma = np.linspace(0,5,100)
cdf = norm.cdf(Nsigma)

fig, ax = plt.subplots(figsize=(10,6))
ax.set_xlabel('Factors of $\sigma$')
ax.set_ylabel('Probability of measurement')
ax.plot(Nsigma, cdf, lw=2)
ax.set_ylim([0,1.1])
plt.show()
```

68% of measurements will fall within $\bar{w} \pm \sigma$, 95% will fall within $\bar{w} \pm 2\sigma$, 99% will fall within $\bar{w} \pm 3\sigma$, and so on asymptoting to unity. That means the probability of taking measurements that fall more than $8\sigma$ from the mean, for example, is highly unlikely. How unlikely? 

```python
print 'Probability of a measurement being more than 8-sigma from the mean = %.15f%%' % (100.0*(1-norm.cdf(8)))
```

# "Statistical Significance"

Let's say you were trying to detect something new with your measurements -- maybe a weird new feature of physics that predicts that the weight of the bucket at a particular time should change by a large amount, and sure enough, you see the weight jump up to 555 grams in one measurement. 

We can discuss how confident our detection of the new phenomenon is using this language of sigmas. If all of our measurements have the same intrinsic uncertainty characterized by the standard deviation of the large group of measurements above, we're claiming a detection of an event like this:

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.errorbar(0, 555, yerr=np.std(bucketweight), fmt='.', color='k', label='$w_i$')
ax.axhline(np.mean(bucketweight), color='k', ls='--', label=r'$\bar{w}$')
for coeff, label in zip([1, -1], [r'$\bar{w}\pm\sigma$', '']):
    ax.axhline(np.mean(bucketweight) + coeff*np.std(bucketweight), color='r', ls='--', label=label)
ax.set_xlabel('Measurement Trial ($i$)')
ax.set_ylabel('Bucket Weight (grams)')
ax.set_title('Bucket weight - no rain')
ax.set_xticks([0])
ax.legend(numpoints=1)
plt.show()
```

The error bars on the point represent the $\pm 1\sigma$ uncertainty on the measurement. 

That point really jumps far outside of where the majority of our measurements were for the bucket before. What's the probability that we'd measure a point to be this far away from the mean? We can take the difference between the measurement and the mean $\bar{w}-w_i$ and divide that by the standard deviation $\sigma$ to find the probability of taking a measurement way out there: 

```python
Nsigma = (555 - np.mean(bucketweight))/np.std(bucketweight)
probability = 100*(1 - norm.cdf(Nsigma))
print 'Number of sigma from mean = %f' % Nsigma
print 'Probability of getting this measurement do to random error = %.6f%%' % probability
```

This new measurement is *unlikely* to be due to the random uncertainty inherent in our weight measurements, since it is $>5\sigma$ from the mean. We'd expect only one in 1.7 milion measurements to fall this far from the mean, so if we have an alternative explanation that says there is some new physical phenomenon that would change the weight of our bucket, perhaps that explanation would be more likely than random error.

We'd call the deviation from the expected value (the mean) in this measurement to be **statistically significant** - meaning our statistics indicate that it is unlikely that the deviation from expectation at this point is large and cannot be accounted for by random error. This is an important idea because it allows us to rely on our stastics to tell us if our data are trying to tell us something (when we know our uncertainties well). If we measure an interesting phenomenon at the $2\sigma$ level, we generally say that the result is *not significant* because there's a decent probability that random uncertainties will produce that measurement. If you measure a $3\sigma$ event, it's more likely a true detection of something interesting, and if you can get measurements at $>5\sigma$ you're even very very likely seeing something interesting and the result is significant.

As an example, physicists were searching for a signal with the Large Hadron Collider at CERN that would signify the detection of a theoretically predicted particle called the [Higgs Boson](http://en.wikipedia.org/wiki/Search_for_the_Higgs_boson). When the LHC detected the new particle at the $5\sigma$ level, [the team announced the official discovery of the new particle](http://press.web.cern.ch/press-releases/2012/07/cern-experiments-observe-particle-consistent-long-sought-higgs-boson), saying they were certain enough that the detection was real and not due to random noise in their data.

# How good is my model?

Now let's return to the original bucket experiment. We're going to measure the flux of raindrops into the bucket my measuring the weight of our bucket over time as it sits out in the rain. Uncertainty on each measurement prevents us from knowing exactly what the weight was at a given time, but it gives us an idea. Here are our data:

```python
time = np.arange(10)
weight = 500 + 10*np.arange(10) + 10*np.random.randn(10) 
errors = np.zeros_like(weight) + 10
# Plot the data
fig, ax = plt.subplots(figsize=(10,6))
ax.errorbar(time, weight, yerr=errors, fmt='o', color='k')
ax.set_xlabel('Time (minutes)')
ax.set_ylabel('Weight (grams)')
ax.set_title('Bucket weight - during rain')
ax.set_xlim([-1,10])
plt.show()
```

First let's do a sanity check -- at $t=0$ we expect the bucket to have it's normal weight, which was 500 grams. The measurement is not exactly 500, but the measurement is "within a few sigma" of the expected value, so we recognize that these measurements make sense so far. Then we see that the weight of the bucket appears to be gradually increasing.

One thing we might want to use our statistics to ask is: "what is the probabilty that we'd get a curve like this if it wasn't raining?" If it was not raining, we'd expect the weight to be equal to the ordinary weight of the bucket at all times. Let's call this the "constant empty bucket hypothesis." Graphically, the expectation is like this:

```python
expectation = np.zeros_like(weight) + 500
fig, ax = plt.subplots(figsize=(10,6))
ax.errorbar(time, weight, yerr=errors, fmt='o', color='k', label='$w_i$')
ax.plot(time, expectation, 'b', lw=2, ls='--', label='Expectation')
ax.legend(numpoints=1, loc='upper left')
ax.set_xlabel('Time (minutes)')
ax.set_ylabel('Weight (grams)')
ax.set_title('Bucket weight - during rain')
ax.set_xlim([-1,10])
plt.show()
```

We can see that at $t=9$ minutes, the point is "many sigma" from the expectation line, so it's probably unlikely that the "constant empty bucket hypothesis" is true.

What we'd really like to do is compare the distance of each point to the expectation, relative to the uncertainty of each point. We do that with a tool called the $\chi^2$ (pronounced "k- eye squared", sometimes written as "chi-squared"). We'll define it as: 

$$ \chi^2 = \sum_{i=1}^N \left( \frac{w_i - f(t_i)}{u_i} \right)^2 $$

where $u_i$ is the uncertainty associated with each weight measurement $w_i$, $N$ is the number of measurements, and $f(t)$ is the expectation function evaluated at of time of the measurement $w_i$. Similarly to the standard deviation, the $\chi^2$ measures how far points fall from something. The standard deviation compares how far the points fall from the mean, the $\chi^2$ compares how far points fall from some arbitrary model $f(t)$ -- which we can choose. Let's take a function like the one plotted above, which represents the "constant empty bucket hypothesis":

$$f(t) = 500 \text{grams}$$ 

The $\chi^2$ for this model is: 

```python
def chisquared(data, model, uncertainties):
    return np.sum( ((data-model)/uncertainties)**2 )

print 'Constant empty bucket model chi^2 = %f' % chisquared(weight, expectation, errors)
```

Now what does that number mean? **The reason $\chi^2$ is useful comes from the fact that a good model has a small $\chi^2$ (meaning small $\chi^2 = $ smaller scatter about the expectation model).**

Now let's compare that empty model to a better model. Using your eyes, you might imagine that a linear function describes the trend well. The intercept looks like it could be about 500 and the slope might be $\Delta y/\Delta x \approx 100/10 = 10$. Our linear model has the form: 

$f(t) = 10t + 500$

Let's over plot this linear model on the plot above:

```python
def linearfunction(times):
    return 10*times + 500

constantmodel = expectation
linearmodel = linearfunction(time)

fig, ax = plt.subplots(figsize=(10,6))
ax.errorbar(time, weight, yerr=errors, fmt='o', color='k', label='$w_i$')
ax.plot(time, expectation, 'b', lw=2, ls='--', label='Constant model')
ax.plot(time, linearmodel, 'r', lw=2, ls='--', label='Linear model')
ax.legend(numpoints=1, loc='upper left')
ax.set_xlabel('Time (minutes)')
ax.set_ylabel('Weight (grams)')
ax.set_title('Bucket weight - during rain')
ax.set_xlim([-1,10])
plt.show()
```

That model looks like it describes the data much better, and so we'd expect a *smaller* $\chi^2$. We get:

```python
print 'Linear model chi^2 = %f' % chisquared(weight, linearmodel, errors)
```

which is significantly smaller! In words, we just showed that the scatter about the linear model is small, and when compared to the much larger $\chi^2$ from the constant empty model, we showed that *the linear model is preferred over the constant empty model*. How significant is the difference?

You can discuss the confidence of this detection using $\sqrt{\Delta\chi^2}$ as a proxy for a "factor of sigma", and then describe the signficance in the difference between the two models using your intuition for sigmas just like we did earlier. In this example: 

```python
deltachi2 = chisquared(weight, expectation, errors) - chisquared(weight, linearmodel, errors)
print "sqrt(Delta chi^2) = %f" % np.sqrt(deltachi2)
```

Here, we've found that the linear model is preferred over the constant empty model at the $\approx 16\sigma$ level, which is a very very significant preference.
